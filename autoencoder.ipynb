```
import tensorflow as tf
```

Encoder

```
class encoder(tf.keras.layers.Layer):
  def __init__(self, dims):
    super(encoder, self).__init__()
    self.hidden = tf.keras.layers.Dense(units = dims, activation = tf.nn.relu)
    self.output = tf.keras.layers.Dense(units = dims, activation = tf.nn.relu)
    
  def call(self, input):
    activation = self.hidden(input)
    return self.output_layer(activation)
```

Decoder

```
class decoder(tf.keras.layers.Layer):
  def __init__(self, dims, orig_dims):
    super(decoder, self).__init__()
    self.hidden = tf.keras.layers.Dense(units = dims, activation = tf.nn.relu)
    self.output = tf.keras.layers.Dense(units = orig_dims, activation = tf.nn.relu)
    
  def call(self, input):
    activation = self.hidden(input)
    return self.output_layer(activation)
```

Putting the encoder and decoder together for the autoencoder

```
class autoencoder(tf.keras.Model):
  def __init__(self, dims, orig_dims):
    super(autoencoder, self).__init__()
    self.encoder = encoder(dims = dims)
    self.decoder = decoder(dims = dims, orig_dims = orig_dims)
    
  def call(self, input):
    encoded = self.encoder(input)
    reconstructed = self.decoder(encoded)
    return reconstructed
```

Loss function

```
def loss(model, orig):
  error = tf.reduce_mean(tf.square(tf.subtract(model(orig), orig)))
  return error
```

Training function

```
def train(loss, model, optim, orig):
  with tf.GradientTape() as tape:
    grads = tape.gradient(loss(model, original), model.trainable_variables)
    grad_variables = zip(grads, model.trainable_variables)
    optim.apply_gradients(grad_variables)
```
